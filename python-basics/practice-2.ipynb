```python
import matplotlib.pyplot as plt

x_data = [1, 2, 3, 4, 5]
y_data = [2, 3, 5, 7, 11]
plt.plot(x_data, y_data, marker='o')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.grid(True)
plt.title('Sample Plot')
```

```python
y_data2 = [1, 4, 6, 8, 10]

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.grid(True)
plt.title('Sample Plot')
plt.plot(x_data, y_data, marker='o')
plt.plot(x_data, y_data2, marker='x', linestyle='--')
plt.legend(['Data 1', 'Data 2'])
```

```python
#bar chart

x_data = ["New York","Paris"]
y_data = [200,233]

plt.bar(x_data,y_data, color=['blue','orange'])
plt.show()
```

```python
# historgram 
f = open('agedata.csv','r')
ages =  f.readlines()

ages_list = []

for age in ages:
    ages_list.append(int(age))
print(ages_list)

bins = [0,1,2,4,20,30,40,50,60,70]
plt.hist(ages_list,bins=bins,histtype='bar',rwidth=0.9)
```

```python
#box plot
plt.title('box plot')

plt.boxplot(ages_list)
```

```python
#pie chart
f = open('agedata2.csv', 'r')

city_list = []

agesdata = f.readlines()
for agedata in agesdata:
    age, city = agedata.split(',')
    city_list.append(city)


from collections import Counter
city_count =Counter(city_list)
city_names = list(city_count.keys())
city_values = list(city_count.values())
print(city_count)
plt.pie(city_values,labels=city_names, autopct="%.3f%%")
```

```python
#scatter plot

f = open('salesdata2.csv','r')

salesdata = f.readlines()

s_list = []
c_list = []

for record  in salesdata:
    sales, cost = record.split(sep=',')
    s_list.append(int(sales))
    c_list.append(int(cost))

plt.scatter(s_list,c_list)
```

```python
# by deafult all plt are on same figure
# to change to new figure and save 
plt.figure('barplot')

plt.scatter(s_list,c_list)
plt.savefig('scatter.png')
```

```python
# subplots
plt.subplot(2,2,1)

plt.scatter(s_list,c_list)
plt.subplot(2,2,2)

plt.pie(city_values,labels=city_names, autopct="%.3f%%")
```
