Handling Missing values

```python
# drop row with missing values 
cleandata =  dataset.dropna()

print(cleandata.shape)
print(cleandata)

# drop column with missing values 
cleandata =  dataset.dropna(axis=1)

print(cleandata.shape)
print(cleandata)
```

```python
# replace na values with mode (most freq value) of given columns
dt =  dataset.copy()
cols = ['Gender','Area','Loan_Status']
dt[cols] = dt[cols].fillna(dt.mode().iloc[0])
dt
```

```python
cols2 = ['LoanAmount']
dt[cols2] = dt[cols2].fillna(float(dt.mean()))
dt
```

Label Encoding of Data
Gender column having male and female  replace them labeled valued 0 and 1


```python
# categorial data to numerical data
# label encoding
print(dt.dtypes)

cols = ['Gender','Area','Loan_Status']
dt[cols] = dt[cols].astype('category')
print(dt.dtypes)

for column in cols:
    dt[column] = dt[column].cat.codes

print(dt)
```

# Normalization
It is means of adjusting values measured on different scales to notionally common scale, often prior to averging 
why? -->
linear regression y =  a x1 + b x2 + c
if x1 values vary from 0 to 10 and x2 from 1000 to 10k then we can say that x2 influence is more on the y than x1

to avoid this problem we by normalizing the x1 and x2 to common scale.
it is required in case of neural network.

1. Standardise the range independent of features or varibles
2. varibles are fitted within certain range
3. applied on alpha numerics values


### normalization methods
1. Z-score (Standardization) Formula: 
z = (x - μ) / σ 
Result: mean ≈ 0, std ≈ 1 Use when: data roughly Gaussian, algorithms assuming centered features (LR, SVM, NN) Pros: handles different scales Cons: affected by outliers (μ, σ shift)

2. Min-Max Scaling Formula (0–1 range): 
x' = (x - min) / (max - min) General range [a,b]: x' = a + ( (x - min)/(max - min) ) * (b - a) 
Result: values in fixed interval Use when: need bounded inputs (distance-based models, NN with sigmoid) Cons: very sensitive to outliers (range stretches)

3. Logistic (Sigmoid) Scaling Basic: 
x' = 1 / (1 + e^{-x}) Often: x standardized first: x_std = (x - μ)/σ, then x' = 1/(1 + e^{-x_std}) 
Result: maps to (0,1) with heavy compression at extremes Use when: want to squash large magnitude differences Cons: loses linear distance info at tails, not invertible for exact original values

Quick comparison

Preserve relative spacing: Z-score (yes, linear), Min-Max (yes within original range), Logistic (no tails compressed)
Bounded: Min-Max (custom), Logistic (0,1), Z-score (unbounded)
Outlier robustness: none (all sensitive; consider robust scaler)
